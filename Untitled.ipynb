```python
"""
Breadth-first search Navigator

Created Oct  4 21:20:15 2024

@author: I_Osunmakinde
"""


#import libraries
import networkx as netX
import matplotlib.pyplot as vis

class BFS_Navigator:   
    #Build network using adjacency matrix            
    def build_network(self):                   
        self.network = {'Main server': ['Mail server', 'Print server'],
                       'Mail server': ['External storage', 'Xerox printer'],
                       'Print server': ['Scanner', 'Hp printer'],
                       'External storage': [],
                       'Xerox printer': [],
                       'Scanner': [],
                       'Hp printer' : []}
                      
    def visualize_Network(self): 
        net = netX.Graph()
        net.add_edge('Main server', 'Mail server')          
        net.add_edge('Main server', 'Print server') 
        
        net.add_edge('Mail server', 'External storage') 
        net.add_edge('Mail server', 'Xerox printer')       
        
        net.add_edge('Print server', 'Scanner')      
        net.add_edge('Print server', 'Hp printer')   
        
        netX.draw(net, with_labels=True)  
        vis.show()
     
    def BF_Search(self, vertex, destination): 
        if destination not in self.network:
            return f"Device '{destination}' does not exist in the network."
                  
        explored = []           # Initalize 
          
        FIFO_queue = []          
        FIFO_queue.append(vertex) 
         
        while FIFO_queue:               
            vertex = FIFO_queue.pop(0)    #shallowest vertex
            
            if vertex not in explored:
               explored.append(vertex)
               if vertex == destination:
                  bfs.visualize_explored_path(explored)
                  return explored
              
               frontiers = self.network[vertex]                
               for frontier in frontiers:
                    FIFO_queue.append(frontier)                   
                     
        return "Path not found"
    
    def visualize_explored_path(self, path):
        net = netX.Graph()
        for vertex in self.network:
            for neighbor in self.network[vertex]:
                net.add_edge(vertex, neighbor)
        
        # Highlight the explored path in red
        path_edges = [(path[i], path[i+1]) for i in range(len(path)-1)]
        pos = netX.spring_layout(net)
        netX.draw(net, pos, with_labels=True, edge_color="black")
        netX.draw_networkx_edges(net, pos, edgelist=path_edges, edge_color="red", width=2)
        vis.show()




# -----------Driving the above class ----------------------------------- 

print('N A V I G A T O R: Breadth-first Search')        

bfs = BFS_Navigator() 
#Build Network
bfs.build_network() 

#Visualize network
bfs.visualize_Network()

# Searching from initial vertex n and display path
initial_state = 'Main server'
print('Root State = ', initial_state)
destination = str(input("Enter Device name: "))

explored = bfs.BF_Search(initial_state, destination) 
print('\n Network Path found from: ', initial_state, ' to: ', destination)
print(explored)



```
