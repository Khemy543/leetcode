```python
"""
Created Oct  2024

@author: I_Osunmakinde
"""

# Depth-first search Navigator + Computational complexity

#import libraries
import networkx as netX
import matplotlib.pyplot as vis
import sys
import time
import memory_profiler as mem_profile

class DFS_Navigator:   
    #Build network using adjacency matrix            
    def build_network(self): 
        self.network = {'Arad1': ['Sibiu', 'Timisoara', 'Zerind'],
                       'Sibiu': ['Arad2', 'Fagaras', 'Oradea1', 'Rimnicu-Vilcea'],
                   'Timisoara': ['Arad3', 'Lugoj'],
                      'Zerind': ['Arad4', 'Oradea2'],
              'Rimnicu-Vilcea': [],
                     'Oradea1' : [], 
                     'Fagaras': [],   
                       'Arad2': [],                                  
                       'Arad3': [],
                      'Lugoj' : [],
                      'Arad4' : [],
                     'Oradea2': []}
        
    def visualize_Network(self): 
        net = netX.Graph()
        net.add_edge('Arad1', 'Sibiu')          
        net.add_edge('Arad1', 'Timisoara') 
        net.add_edge('Arad1', 'Zerind') 
        
        net.add_edge('Sibiu', 'Arad2') 
        net.add_edge('Sibiu', 'Fagaras') 
        net.add_edge('Sibiu', 'Oradea1')      
        net.add_edge('Sibiu', 'Rimnicu-Vilcea')        
        
        net.add_edge('Timisoara', 'Arad3')      
        net.add_edge('Timisoara', 'Lugoj')   
        
        net.add_edge('Zerind', 'Arad4')      
        net.add_edge('Zerind', 'Oradea2')   
        
        netX.draw(net, with_labels=True)  
        vis.show() 
                    
    def visualize_explored_path(self, path):
        net = netX.Graph()
        for vertex in self.network:
            for neighbor in self.network[vertex]:
                net.add_edge(vertex, neighbor)

        print(net)
        # Highlight the explored path in red
        path_edges = [(path[i], path[i+1]) for i in range(len(path)-1)]
        pos = netX.spring_layout(net)
        netX.draw(net, pos, with_labels=True, edge_color="black")
        netX.draw_networkx_edges(net, pos, edgelist=path_edges, edge_color="red", width=2)
        vis.show()
            
    def DF_Search(self, explored, vertex, destination):   
                
        if vertex not in explored:
            print (vertex)
            explored.append(vertex)
            
            #compare with destination
            if vertex == destination:        
                #End Time and memory complexity
                
                t2 = time.process_time()
                print('\n Time complexity = ', abs(t2-t1), ' seconds')      #Time
                
                m2 = mem_profile.memory_usage()
                print('\n Memory complexity = ', m2[0] - m1[0], ' Mb')      #Memory  

                #highlight explored path
                self.visualize_explored_path(explored)
                exit()
                
            else:
                frontiers = self.network[vertex]                
                for neighbour in frontiers:
                    self.DF_Search(explored, neighbour, destination)   #recursion
                    
                 
                    
                 
                    
        
# -----------Driving the above class ----------------------------------- 

print('N A V I G A T O R: Depth-first Search')

#Start Time and memory complexity           
t1 = time.process_time()                    #Time
m1 = mem_profile.memory_usage()             #Memory

dfs = DFS_Navigator() 
dfs.build_network()

dfs.visualize_Network()


# Searching from vertex n and display path
initial_state = 'Arad1'
print('Root State = ', initial_state)
destination = str(input("Enter Destination State: "))

print('\n Path found from: ', initial_state, ' to: ', destination)

explored = list()           # set keeps track of explored vertex
dfs.DF_Search(explored, initial_state, destination)
```
